```python
import pywt
import pywt.data
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = [15, 7]

x = pywt.data.ecg()
x.shape
```

```python
cA1, cD1 = pywt.dwt(x, 'db1')
t = np.linspace(0,511,512)
plt.plot(t, cA1)
```

```python
cA2, cD2 = pywt.dwt(cA1, 'db1')
t2 = np.linspace(0, 255, 256)
plt.plot(t2, cA2)
```

```python
cA3, cD3 = pywt.dwt(cA2, 'db1')
t3 = np.linspace(0, 127, 128)
plt.plot(t3, cA3)
```
